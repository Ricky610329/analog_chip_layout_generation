```python
from aclg.rules.split.split_ratio import split_by_ratio, SplitOrientation, split_by_ratio_grid
from aclg.rules.spacing import spacing_grid, spacing_vertical, spacing_horizontal
from aclg.post_processing.padding import add_padding
from aclg.rules.symetric.symmetric_1 import split_symmetric_1_horizontal, split_symmetric_1_vertical
from aclg.rules.align import align_components, AlignmentMode
from aclg.dataclass.component import Component
import random

number_of_instance = 10
result = []

for i in range(number_of_instance):
    relation_id = 0
    level = 0
    component_hierarchy = {}

    # level 0
    # generate root
    root_component = Component(
        x=0,
        y=0,
        width=random.randint(50, 130),
        height=random.randint(50, 130),
        relation_id=relation_id,
        generate_rule='root',
        level=level
    )
    component_hierarchy['level_0'] = [root_component]
    relation_id += 1
    level += 1

    # level 1
    # align only level
    choose_align_rule = random.choice([AlignmentMode.LEFT, AlignmentMode.RIGHT, AlignmentMode.TOP, AlignmentMode.BOTTOM, AlignmentMode.CENTER_H, AlignmentMode.CENTER_V])
    split_count = random.randint(2, 5)
    ratios = [random.randint(1, 3) for _ in range(split_count)]
    scaling_factors = [random.uniform(0.3, 0.8) for _ in range(split_count)]
    level_1_components = []
    if choose_align_rule in [AlignmentMode.LEFT, AlignmentMode.RIGHT, AlignmentMode.CENTER_V]:
        level_1_components = split_by_ratio(component=root_component, ratios=ratios, orientation=SplitOrientation.HORIZONTAL)
    elif choose_align_rule in [AlignmentMode.TOP, AlignmentMode.BOTTOM, AlignmentMode.CENTER_H]:
        level_1_components = split_by_ratio(component=root_component, ratios=ratios, orientation=SplitOrientation.VERTICAL)

    level_1_components = align_components(components=level_1_components, scale_factors=scaling_factors, mode=choose_align_rule)

    for comp in level_1_components:
        comp.relation_id = relation_id
        comp.level = level
    relation_id += 1
    level += 1

    root_component.sub_components = level_1_components
    component_hierarchy['level_1'] = level_1_components

    # level 2
    for comp in level_1_components:
        choose_rule = random.choice(['split_by_ratio', 'symmetric', 'spacing'])
        split_count = random.randint(2, 3)
        level_2_components = []
        if choose_rule == 'split_by_ratio':
            ratios = [random.randint(1, 5) for _ in range(split_count)]
            level_2_components = split_by_ratio(component=comp, ratios=ratios,
                                                orientation=random.choice([SplitOrientation.HORIZONTAL, SplitOrientation.VERTICAL]))
        elif choose_rule == 'symmetric':
            symmetric_function = random.choice([split_symmetric_1_horizontal, split_symmetric_1_vertical])
            level_2_components = symmetric_function(component=comp)
        elif choose_rule == 'spacing':
            spacing_function = random.choice([spacing_vertical, spacing_horizontal])
            level_2_components = spacing_function(component=comp, num_components=split_count)

        for comp in level_2_components:
            comp.relation_id = relation_id
            comp.level = level
        relation_id += 1

        comp.sub_components = level_2_components
        component_hierarchy[f'level_2_{comp.relation_id}'] = level_2_components

    result.append(component_hierarchy)
```

```python
# Pillow library for drawing
from PIL import Image, ImageDraw
import random

# aclg imports from the user's provided code
from aclg.rules.split.split_ratio import split_by_ratio, SplitOrientation, split_by_ratio_grid
from aclg.rules.spacing import spacing_grid, spacing_vertical, spacing_horizontal
from aclg.rules.symetric.symmetric_1 import split_symmetric_1_horizontal, split_symmetric_1_vertical
from aclg.rules.align import align_components, AlignmentMode
from aclg.dataclass.component import Component


# --- Drawing Settings ---
CANVAS_SIZE = (300, 300) # A larger canvas for potentially larger layouts
OFFSET = (CANVAS_SIZE[0] / 2, CANVAS_SIZE[1] / 2) # Center point
number_of_instance = 10
result = []

# --- Color Scheme for Visualization ---
# Using RGBA tuples (Red, Green, Blue, Alpha/Transparency)
COLOR_LEVEL_0 = (100, 100, 100, 120) # Semi-transparent grey
OUTLINE_LEVEL_0 = 'black'
COLOR_LEVEL_1 = (60, 120, 255, 150)  # Semi-transparent blue
OUTLINE_LEVEL_1 = (0, 0, 139)       # Dark blue
COLOR_LEVEL_2 = (255, 60, 30, 180)  # Semi-transparent red
OUTLINE_LEVEL_2 = (139, 0, 0)       # Dark red

print(f"[*] Starting generation of {number_of_instance} three-level layout images...")

for i in range(number_of_instance):
    # --- Generation Logic ---
    relation_id = 0
    level = 0
    component_hierarchy = {}

    # level 0
    root_component = Component(
        x=0, y=0, width=random.randint(50, 130), height=random.randint(50, 130),
        relation_id=relation_id, generate_rule='root', level=level
    )
    component_hierarchy['level_0'] = [root_component]
    relation_id += 1
    level += 1

    # level 1
    choose_align_rule = random.choice(list(AlignmentMode))
    split_count = random.randint(2, 5)
    ratios = [random.randint(1, 3) for _ in range(split_count)]
    scaling_factors = [random.uniform(0.3, 0.8) for _ in range(split_count)]
    level_1_components = []
    if choose_align_rule in [AlignmentMode.LEFT, AlignmentMode.RIGHT, AlignmentMode.CENTER_V]:
        level_1_components = split_by_ratio(component=root_component, ratios=ratios, orientation=SplitOrientation.HORIZONTAL)
    elif choose_align_rule in [AlignmentMode.TOP, AlignmentMode.BOTTOM, AlignmentMode.CENTER_H]:
        level_1_components = split_by_ratio(component=root_component, ratios=ratios, orientation=SplitOrientation.VERTICAL)

    level_1_components = align_components(components=level_1_components, scale_factors=scaling_factors, mode=choose_align_rule)
    for comp in level_1_components:
        comp.relation_id = relation_id
        comp.level = level
    relation_id += 1
    level += 1
    root_component.sub_components = level_1_components
    component_hierarchy['level_1'] = level_1_components

    # level 2
    parent_comp_index = 0
    for parent_comp in level_1_components:
        choose_rule = random.choice(['spacing_grid', 'split_by_ratio_grid'])
        level_2_components = []
        if choose_rule == 'spacing_grid':
            rows = random.randint(1, 2)
            cols = random.randint(1, 2)
            level_2_components = spacing_grid(component=parent_comp, rows=rows, cols=cols)
        elif choose_rule == 'split_by_ratio_grid':
            num_h = random.randint(1, 2)
            num_v = random.randint(1, 2)
            h_ratios = [random.uniform(0.5, 1.5) for _ in range(num_h)]
            v_ratios = [random.uniform(0.5, 1.5) for _ in range(num_v)]
            level_2_components = split_by_ratio_grid(component=parent_comp, h_ratios=h_ratios, v_ratios=v_ratios)

        current_relation_id = relation_id + parent_comp_index
        for comp in level_2_components:
            comp.relation_id = current_relation_id
            comp.level = level
        
        parent_comp.sub_components = level_2_components
        component_hierarchy[f'level_2_rel_{current_relation_id}'] = level_2_components
        parent_comp_index += 1
    
    relation_id += parent_comp_index
    result.append(component_hierarchy)

    # --- Start Drawing ---
    img = Image.new('RGBA', CANVAS_SIZE, (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)

    # 1. Draw Level 0 (Root)
    root = component_hierarchy['level_0'][0]
    tl_x, tl_y = root.get_topleft()
    br_x, br_y = root.get_bottomright()
    draw.rectangle(
        [(tl_x + OFFSET[0], tl_y + OFFSET[1]), (br_x + OFFSET[0], br_y + OFFSET[1])],
        fill=COLOR_LEVEL_0, outline=OUTLINE_LEVEL_0, width=2
    )
    
    # 2. Draw Level 1 Components
    for component in component_hierarchy['level_1']:
        tl_x, tl_y = component.get_topleft()
        br_x, br_y = component.get_bottomright()
        draw.rectangle(
            [(tl_x + OFFSET[0], tl_y + OFFSET[1]), (br_x + OFFSET[0], br_y + OFFSET[1])],
            fill=COLOR_LEVEL_1, outline=OUTLINE_LEVEL_1, width=1
        )

    # 3. Draw Level 2 Components
    # Iterate through the dictionary to find all level 2 keys
    for key, components in component_hierarchy.items():
        if key.startswith('level_2_'):
            for component in components:
                tl_x, tl_y = component.get_topleft()
                br_x, br_y = component.get_bottomright()
                draw.rectangle(
                    [(tl_x + OFFSET[0], tl_y + OFFSET[1]), (br_x + OFFSET[0], br_y + OFFSET[1])],
                    fill=COLOR_LEVEL_2, outline=OUTLINE_LEVEL_2, width=1
                )

    # 4. Save the final image
    filename = f"3level_hierarchy_{i+1}.png"
    img.save(filename)
    print(f"  -> Saved image: {filename}")

print("\n[+] Success! All images have been generated. ✨")
```
